```python
df3=df.groupby("Prime_Minister")[["GDP_Beginning","GDP_End"]].max()
```

```python
from matplotlib.ticker import FuncFormatter
colors = ["#a1d99b", "#e5f5e0"]
pm = df["Prime_Minister"].unique()
df01= df3.reindex(pm)
ax = df01.plot(kind="bar", figsize=(18,9),color=colors)
plt.ylabel("GDP (in crores)")
plt.title("GDP at Beginning vs End of Tenure")
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x/1000:.0f}'))
ax.set_yscale('log')
for container in ax.containers:
    ax.bar_label(container,
        labels=[f'{int(v):,}' for v in container.datavalues],
        padding=3,
        fontsize=8)
```

```python
df3["GDP_Percentage_Increase"] = (
    (df3["GDP_End"] - df3["GDP_Beginning"]) / df3["GDP_Beginning"]) * 100
df3["GDP_Percentage_Increase"] = (
    (df3["GDP_End"] - df3["GDP_Beginning"]) / df3["GDP_Beginning"]) * 100
colors=[  "#e5f5e0",  "#c7e9c0","#a1d99b","#74c476","#41ab5d","#238b45","#006d2c"]
df_positive = df3[df3["GDP_Percentage_Increase"] > 0]
top5 = df_positive.sort_values(
    by="GDP_Percentage_Increase",
    ascending=False).head(7)
plt.figure(figsize=(6,6))
plt.pie(top5["GDP_Percentage_Increase"],labels=top5.index,autopct='%1.1f%%',startangle=140,colors=colors)
plt.title("GDP Percentage Increase")
plt.show()
```

```python
plt.figure(figsize=(8,0.1))
plt.text(0.2,0.20, "POPULATION GROWTH:", fontsize=20, fontweight='bold', ha='center', va='top')
plt.axis("off")
plt.show()
```

```python
df4=df.groupby("Prime_Minister")[["Population_at_Beginning","Population_at_End"]].max()
```

```python
pm_order = df["Prime_Minister"].unique()
df4 = df4.reindex(pm_order)
colors=["#bbdefb","#42a5f5"]
df5=df4.plot( kind="bar", figsize=(18,8),color=colors)
plt.ylabel("population(in millions)")
plt.title("Population at Beginning vs End of Tenure")
for container in df5.containers:
    df5.bar_label(container, fmt='%.1f', padding=3)
```

```python
df4["Population_Percentage_Increase"] = (
    (df4["Population_at_End"] - df4["Population_at_Beginning"])/ df4["Population_at_Beginning"]) * 100
df4_pos = df4[df4["Population_Percentage_Increase"] > 0] \
            .sort_values("Population_Percentage_Increase", ascending=False) \
            .head(7)
colors=[ "#e3f2fd","#bbdefb","#90caf9","#64b5f6","#42a5f5","#1e88e5","#1565c0","#0d47a1","#082567"]
explode = [0.06] * len(df4_pos)
plt.figure(figsize=(6,6))
plt.pie(df4_pos["Population_Percentage_Increase"],labels=df4_pos.index,autopct='%1.1f%%',explode=explode,startangle=140,colors=colors)
plt.title("Population Percentage Increase During Tenure")
plt.show()
```

```python
df["GDP per Capita Avg"] = (df["GDP per Capita at Beginning of Tenure (in US $)"] + 
                            df["GDP per Capita at End"]) / 2
df["Population Growth Avg"] = (df["Population Growth at Beginning of Tenure (in % per annum)"] + 
                               df["Population Growth at End of Tenure (in % per annum)"]) / 2
```

```python
fig, ax1 = plt.subplots(figsize=(21,10))

ax1.plot(df["Prime_Minister"], df["GDP per Capita Avg"], color="green", marker="o", label="GDP per Capita (US $)")
ax1.set_ylabel("GDP per Capita (US $)", color="blue")
ax2 = ax1.twinx()
ax2.plot(df["Prime_Minister"], df["Population Growth Avg"], color="blue", marker="s", label="Population Growth (%)")
ax2.set_ylabel("Population Growth (% per annum)", color="green")
plt.title("GDP per Capita vs Population Growth",fontsize=20)
plt.grid(True)
fig.tight_layout()
fig.legend(loc="upper right")    
plt.xticks(rotation=90)
plt.show()    #inverse proportion
```

```python
df["Income_Stage"] = df["GDP per Capita at End"].apply(
    lambda x: "Low-income" if x < 1145 else "Lower-middle" if x < 4515 else "Upper-middle" if x < 14005 else "High-income")
```

```python
plt.figure(figsize=(15,8))
bars = plt.bar(df["Prime_Minister"], df["GDP per Capita at End"], color="skyblue")
for bar, stage in zip(bars, df["Income_Stage"]):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 100,
             stage, ha="center", va="bottom", fontsize=9)

plt.title("India’s Economic Stage by GDP per Capita")
plt.xlabel("Prime Minister")
plt.ylabel("GDP per Capita ($)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
